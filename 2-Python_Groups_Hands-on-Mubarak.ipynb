#### Q7: Create 2 functions to sort the books by the title and the year from your library and test it.

```python
# write your code here ^_^
def sort_by_title(b_dict):
    sorted_dict = sorted(b_dict.items(), key=lambda x: x[1]['Title'])
    return dict(sorted_dict)

def sort_by_year(b_dict):
    sorted_dict = sorted(b_dict.items(), key=lambda x: x[1]['Year'])
    return dict(sorted_dict)

d = sort_by_title(books)
d
```

#### Q8: Create a function to bulk update genres of books using list comprehension. and test it.

```python
# write your code here ^_^
def update_genres(library, book_id, old_genres, new_genres):
    if book_id in library:
        genres = library[book_id]['Genres']
        library[book_id]['Genres'] = [new_genres if genre == old_genres else genre for genre in genres]
    else:
        print(f"Book with ID {book_id} does not exist.")
    return library

update_genres(books, 1, "Fiction", "Classics")
books
```
